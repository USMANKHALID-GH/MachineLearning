```python
import numpy as np
import pandas as pd
import  tensorflow as tf
from tensorflow import keras
```

```python

 
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/13_dropout_layer/sonar_dataset.csv', header=None)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head(10)
```

```python
X=df.drop(60, axis=1)
```

```python
y=df[60]
```

```python
X.head(3)
```

```python
y=pd.get_dummies(data=y ,drop_first=True)
```

```python
y.head(3)
```

```python
y.value_counts()
```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.25)
```

```python
X_train.shape
```

```python
# X_train=np.array(X_train)
# y_train=np.array(y_train)
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60, activation="relu"),
     keras.layers.Dense(40, activation="relu"),
       keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(20, activation="relu"),
     keras.layers.Dense(10, activation="relu"),
      keras.layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

```

```python
model.fit(X_train, y_train, epochs=100 , batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60, activation="relu"),
    keras.layers.Dropout(.4),
     keras.layers.Dense(40, activation="relu"),
       keras.layers.Dense(30, activation="relu"),
      keras.layers.Dropout(.4),
    keras.layers.Dense(20, activation="relu"),
  keras.layers.Dropout(.4),
     keras.layers.Dense(10, activation="relu"),
      keras.layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="SGD",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
history=model.fit(X_train, y_train, epochs=100 , batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```
